```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
raining=ctrl.Antecedent(np.arange(0,101,10),'Rain')
raining['no rain'] = fuzzy.trimf(raining.universe,[0,0,0])
raining['rain'] = fuzzy.trimf(raining.universe,[0,50,100])
raining.view()
plt.show()
```

```python
rain_intensity=ctrl.Antecedent(np.arange(0,110,10),'Rain intensity')
rain_intensity['drizzle'] = fuzzy.trimf(rain_intensity.universe,[0,20,30])
rain_intensity['raining'] = fuzzy.trimf(rain_intensity.universe,[20,40,70])
rain_intensity['heavy'] = fuzzy.trimf(rain_intensity.universe,[60,80,100])
rain_intensity.view()
plt.show()
```

```python
wiper_speed=ctrl.Consequent(np.arange(0,110,10),'wiper speed')
wiper_speed['off'] = fuzzy.trimf(wiper_speed.universe,[0,0,0])
wiper_speed['low'] = fuzzy.trimf(wiper_speed.universe,[0,30,60])
wiper_speed['high'] = fuzzy.trimf(wiper_speed.universe,[50,70,100])
wiper_speed.view()
plt.show()
```

```python
rule1=ctrl.Rule(raining['no rain'] & rain_intensity['drizzle'], wiper_speed['off'])
rule2=ctrl.Rule(raining['no rain'] & rain_intensity['raining'], wiper_speed['off'])
rule3=ctrl.Rule(raining['no rain'] & rain_intensity['heavy'], wiper_speed['off'])
rule4=ctrl.Rule(raining['rain'] & rain_intensity['drizzle'],wiper_speed['low'])
rule5=ctrl.Rule(raining['rain'] & rain_intensity['raining'], wiper_speed['low'])
rule6=ctrl.Rule(raining['rain'] & rain_intensity['heavy'], wiper_speed ['high'])
rule=[rule1,rule2,rule3,rule4,rule5,rule6]
```

```python
a=ctrl.ControlSystem(rule)
b=ctrl.ControlSystemSimulation(a)
b.input['Rain'] = 10
b.input['Rain intensity'] = 50
b.compute()
print(b.output['wiper speed'])
wiper_speed.view(sim=b)
plt.show()
```

```python
light_intensity=ctrl.Antecedent(np.arange(0,1.1,0.1),'Light intensity')
light_intensity['Satisfactory'] = fuzzy.trimf(light_intensity.universe,[0,0.1,0.2])
light_intensity['Just Acceptable'] = fuzzy.trimf(light_intensity.universe,[0.2,0.4,0.6])
light_intensity['Unbearable'] = fuzzy.trimf(light_intensity.universe,[0.5,0.8,1])
light_intensity.view()
plt.show()
```

```python
light_mode=ctrl.Consequent(np.arange(0,1.1,0.1),'light mode')
light_mode['low'] = fuzzy.trimf(light_mode.universe,[0,0.4,0.7])
light_mode['high'] = fuzzy.trimf(light_mode.universe,[0.5,0.8,1])
light_mode.view()
plt.show()
```

```python
rule7=ctrl.Rule(raining['no rain'] & rain_intensity['drizzle'] & light_intensity['Satisfactory'], light_mode['high'])
rule8=ctrl.Rule(raining['no rain'] & rain_intensity['drizzle'] & light_intensity['Just Acceptable'], light_mode['high'])
rule9=ctrl.Rule(raining['no rain'] & rain_intensity['drizzle'] & light_intensity['Unbearable'], light_mode['low'])
rule10=ctrl.Rule(raining['no rain'] & rain_intensity['raining'] & light_intensity['Satisfactory'], light_mode['high'])
rule11=ctrl.Rule(raining['no rain'] & rain_intensity['raining'] & light_intensity['Just Acceptable'], light_mode['high'])
rule12=ctrl.Rule(raining['no rain'] & rain_intensity['raining'] & light_intensity['Unbearable'], light_mode['low'])
rule13=ctrl.Rule(raining['no rain'] & rain_intensity['heavy'] & light_intensity['Satisfactory'], light_mode['high'])
rule14=ctrl.Rule(raining['no rain'] & rain_intensity['heavy'] & light_intensity['Just Acceptable'], light_mode['high'])
rule15=ctrl.Rule(raining['no rain'] & rain_intensity['heavy'] & light_intensity['Unbearable'], light_mode['low'])
rule16=ctrl.Rule(raining['rain'] & rain_intensity['drizzle'] & light_intensity['Satisfactory'], light_mode['high'])
rule17=ctrl.Rule(raining['rain'] & rain_intensity['drizzle'] & light_intensity['Just Acceptable'], light_mode['high'])
rule18=ctrl.Rule(raining['rain'] & rain_intensity['drizzle'] & light_intensity['Unbearable'], light_mode['low'])
rule19=ctrl.Rule(raining['rain'] & rain_intensity['raining'] & light_intensity['Satisfactory'], light_mode['high'])
rule20=ctrl.Rule(raining['rain'] & rain_intensity['raining'] & light_intensity['Just Acceptable'], light_mode['low'])
rule21=ctrl.Rule(raining['rain'] & rain_intensity['raining'] & light_intensity['Unbearable'], light_mode['low'])
rule22=ctrl.Rule(raining['rain'] & rain_intensity['heavy'] & light_intensity['Satisfactory'], light_mode['low'])
rule23=ctrl.Rule(raining['rain'] & rain_intensity['heavy'] & light_intensity['Just Acceptable'], light_mode['low'])
rule24=ctrl.Rule(raining['rain'] & rain_intensity['heavy'] & light_intensity['Unbearable'], light_mode['low'])
rul_two=[rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24]
```

```python
c4=ctrl.ControlSystem(rul_two)
d4=ctrl.ControlSystemSimulation(c4)
```

```python
d4.input['Rain'] = 65
d4.input['Rain intensity'] = 89
d4.input['Light intensity'] = 0.3
d4.compute()
print(d4.output['light mode'])
```

```python
light_mode.view(sim=d4)
plt.show()
```
